# Prepocessing and Cleaning

> First we Select numeric columns from the datset 
 to fill the null values on it with mean

## Fill null Values 

```python
# Select numeric columns.

numeric_col = df.select_dtypes('number')
numeric_col
```

```python
# Fill numeric columns with mean.
df[numeric_col.columns] = numeric_col.fillna(numeric_col.mean())
```

```python
#check the null values 
df.isna().sum()
```

```python
# fill Categorical column TailNum with mode 
df['TailNum'] = df['TailNum'].fillna(df['TailNum'].mode()[0])
```

```python
#check the null values 
df.isna().sum()
```

## Map name of Month and DayOfWeek 

```python
#Map Month Name

df['Month']=df['Month'].map({1: 'January', 2: 'Feburary',3:'March',4:'April',5:'May',6:'June',
                             7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'})
```

```python
df['Month'].value_counts()
```

```python
#Map DayOfWeek name

df['DayOfWeek']=df['DayOfWeek'].map({1: 'Monday', 2: 'Tuesday',3:'Wednesday',4:'Thursday',
                                     5:'Friday',6:'Saturday',7:'Sunday'})

```

```python
df['DayOfWeek'].value_counts()
```

```python
df.columns
```

## change datatype 

```python

df = df.astype({"ActualElapsedTime": int}, errors='raise')
df = df.astype({"CRSElapsedTime": int}, errors='raise')
df = df.astype({"AirTime": int}, errors='raise')
df = df.astype({"ArrDelay": int}, errors='raise')
df = df.astype({"DepDelay": int}, errors='raise')
df = df.astype({"TaxiIn": int}, errors='raise')
df = df.astype({"TaxiOut": int}, errors='raise')
df = df.astype({"CarrierDelay": int}, errors='raise')
df = df.astype({"WeatherDelay": int}, errors='raise')
df = df.astype({"NASDelay": int}, errors='raise')
df = df.astype({"SecurityDelay": int}, errors='raise')
df = df.astype({"LateAircraftDelay": int}, errors='raise')

```

```python
df.info()
```

## Drop unwanted column 

```python
#drop uneeded columns
df= df.drop(['Unnamed: 0'], axis=1)
df= df.drop(['Year'], axis=1)
df= df.drop(['DepTime'], axis=1)
df= df.drop(['CRSDepTime'], axis=1)
df= df.drop(['ArrTime'], axis=1)
df= df.drop(['CRSArrTime'], axis=1)
```
